```python
import numpy as np
def generarDistribucionV2(field,x_label,title,*args):
    statField=datos[field].describe()
    print(statField)
    dictField=statField.to_dict()
    dict1=datos[field].to_dict()
    dict2={}
    for key in dict1:
        elemento=dict1[key]
        if elemento!=valorAlternoNan:
            dict2[elemento]=int(0)
        else:
            print(f'se ha detectado un valor invalido:{valorAlternoNan}')
    values=dict1.values()
    for key in dict2:
        dict2[key]=sum(x == key for x in values)
    #print(dict2)
    print('Cantidad de valores encontrados: ',sum(val for val in dict2.values()))
    names=dict2.keys()
    values=dict2.values()
    for tipo in args:
        fig, axs = plt.subplots(1, figsize=(20, 10), sharey=True)
        plt.title(title)
        plt.xlabel(x_label) 
        plt.ylabel('Cantidad Personas')
        if tipo=='scatter':
            axs.scatter(names,values,s=125)
        elif tipo=='bar':
            axs.bar(names,values,width = 0.8)
        elif tipo=='normal':
            create_normal(fig,axs,dictField['mean'],dictField['std'],len(dict2),names)
        elif tipo=='pie':
            plt.pie(values, labels = names,startangle=90, shadow = True, radius = 1.2, autopct = '%1.1f%%') 
        else:
            raise Exception('No se ha reconocido el tipo de grafico indicado')
    if len(tipo)>0:
            plt.show()
            
def create_normal(fig,ax, mu, sigma,num_bins, x):
    count, bins = np.histogram(list(x), num_bins)

    y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
         np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
    ax.plot(bins, y, '-')

    fig.tight_layout()
```
